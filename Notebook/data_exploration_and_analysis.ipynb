# Checked missing values

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
#  Filling numerical missing values with median
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols] = df[numerical_cols].apply(lambda x: x.fillna(x.median()), axis=0)
```

```python
# Filling categorical missing values with 'Unknown'
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.fillna('Unknown'))
```

```python
df['Start'] = df['Start'].fillna('2019-04-29 07:08:38')
```

```python
df['End'] = df['End'].fillna('2019-04-25 00:01:32')
```

```python
df.isnull().sum().iloc[0:55]
```

# Outliers by quantile method

```python
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

outliers = df[(df[numerical_cols] < lower_bound) | (df[numerical_cols] > upper_bound)]
```

# Bosplot for Outliers

```python
plt.figure(figsize=(12,3),dpi=200)
outliers.plot(kind = 'box',color = 'red')
plt.xticks(rotation = 90)
plt.title('total outliers', color = 'red')
plt.ylabel('values',color = 'red')
plt.show()
```
